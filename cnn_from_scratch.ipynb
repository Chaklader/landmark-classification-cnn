- How does your code resize the images (by cropping, stretching, etc)?  What size did you pick for the input tensor, and why?

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> **Answer**: the preprocessing pipeline includes:

1. Image Resizing:

- Images are first resized to 256 pixels using transforms.Resize(256)

- Then center-cropped to 224×224 using transforms.CenterCrop(224)

- This 224×224 size was chosen because it's a standard input size used by many modern CNN architectures (like ResNet) and provides a good balance between preserving image detail and computational efficiency

2. Data Augmentation:

For training data, several augmentation techniques are applied:

- RandomResizedCrop(224): Randomly crops and resizes images to 224×224

- RandomHorizontalFlip(): Randomly flips images horizontally

- RandomRotation(10): Applies random rotations up to 10 degrees

- ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.1): Adjusts color properties

These augmentations help:

- Increase dataset variety

- Improve model generalization

- Make the model more robust to variations in image orientation and lighting conditions

3. Normalization:

- Images are converted to tensors and normalized using computed mean and standard deviation

- This standardizes the input data and helps with training stability

<img src="static_images/icons/noun-question-mark-869751.png" alt="?" style="width:25px"/> **Question**: Outline the steps you took to get to your final CNN architecture and your reasoning at each step.  

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> __Answer:__ The CNN architecture is developed with these considerations:

1. Feature Extractor:

- Uses 4 convolutional blocks with increasing feature depths (64→128→256→512)

- Each block contains:

  - Conv2D layer with 3×3 kernels and padding

  - BatchNormalization for training stability

  - ReLU activation for non-linearity

  - MaxPooling to reduce spatial dimensions

  - Dropout for regularization (increasing from p/2 to p through the network)

2. Classifier:

- Flattens the features

- Uses a two-layer fully connected network:

  - First layer: 512*14*14 → 1024 with ReLU

  - Final layer: 1024 → num_classes

  - Dropout between layers for regularization

The architecture follows modern CNN design principles:

- Gradually increasing feature channels

- Regular dimensionality reduction through pooling

- Heavy regularization (BatchNorm + Dropout)

- Skip connections were not used but could be a potential improvement

The final architecture balances model capacity with computational efficiency while incorporating strong regularization to prevent overfitting.

## <img src="static_images/icons/noun-advance-2109145.png" alt=">" style="width:50px"/> Step 5: Putting everything together

Allright, good job getting here! Now it's time to see if all our hard work pays off. In the following cell we will train your model and validate it against the validation set.

Let's start by defining a few hyperparameters. Feel free to experiment with different values and try to optimize your model:

````python
batch_size = 16        # Even smaller batch size
valid_size = 0.2       
num_epochs = 150       
num_classes = 50       
dropout = 0.2          # Even less dropout if underfitting
learning_rate = 0.0001 # Very conservative learning rate
opt = 'adam'          
weight_decay = 0.00001 # Very light regularization
````

````python
from src.data import get_data_loaders
from src.train import optimize
from src.optimization import get_optimizer, get_loss
from src.model import MyModel
import torch

# get the data loaders using batch_size and valid_size defined in the previous
# cell
# HINT: do NOT copy/paste the values. Use the variables instead
# get the data loaders using batch_size and valid_size 
data_loaders = get_data_loaders(
    batch_size=batch_size,
    valid_size=valid_size
)

# instance model MyModel with num_classes and drouput defined in the previous
# cell
# instance model MyModel with num_classes and dropout
model = MyModel(
    num_classes=num_classes,
    dropout=dropout
)

# Get the optimizer using get_optimizer and the model you just created, the learning rate,
# the optimizer and the weight decay specified in the previous cell
# Get the optimizer using specified parameters
optimizer = get_optimizer(
    model=model,
    optimizer=opt,
    learning_rate=learning_rate,
    weight_decay=weight_decay
)

# Get the loss using get_loss
# Get the loss
loss = get_loss()

optimize(
    data_loaders,
    model,
    optimizer,
    loss,
    n_epochs=num_epochs,
    save_path="checkpoints/best_val_loss.pt",
    interactive_tracking=True
)


# After the optimize function returns the model

# For validation set
def evaluate_model(model, data_loader, device=None):
    if device is None:
        if torch.backends.mps.is_available():
            device = torch.device("mps")
        elif torch.cuda.is_available():
            device = torch.device("cuda")
        else:
            device = torch.device("cpu")
            
    model = model.to(device)
    model.eval()
    
    correct = 0
    total = 0
    
    with torch.no_grad():
        for data, target in data_loader:
            data, target = data.to(device), target.to(device)
            outputs = model(data)
            _, predicted = torch.max(outputs.data, 1)
            total += target.size(0)
            correct += (predicted == target).sum().item()
    
    accuracy = 100 * correct / total
    return accuracy

# Get accuracies
print("\nFinal Model Evaluation:")
print("-" * 40)
val_accuracy = evaluate_model(model, data_loaders['valid'])
test_accuracy = evaluate_model(model, data_loaders['test'])
print(f"Validation Accuracy: {val_accuracy:.2f}%")
print(f"Test Accuracy: {test_accuracy:.2f}%")
````
